```python
fn = 'train.csv'
df = pd.read_csv(fn)
print("Shape:", df.shape)
display(df.head())
print("\nColumns:", list(df.columns)[:80])
print("\nMissing values (top 20):")
print(df.isnull().sum().sort_values(ascending=False).head(20))
```

```python
df_simple = df.copy()

df_simple['Bathrooms'] = df_simple['FullBath'] + 0.5 * df_simple['HalfBath']

features = ['GrLivArea', 'TotalBsmtSF', 'BedroomAbvGr', 'Bathrooms']
target = 'SalePrice'

df_simple = df_simple[features + [target]].dropna(subset=[target])
df_simple[features].isnull().sum()
```

```python
df_simple.describe()

for col in features:
    plt.figure(figsize=(6,4))
    plt.scatter(df_simple[col], df_simple[target], alpha=0.4)
    plt.title(f'{col} vs {target}')
    plt.xlabel(col)
    plt.ylabel(target)
    plt.show()
```

```python
df_simple['TotalBsmtSF'] = df_simple['TotalBsmtSF'].fillna(0)

X = df_simple[features]
y = df_simple[target]

y_log = np.log1p(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_log, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

y_pred_log = lr.predict(X_test_scaled)
y_pred = np.expm1(y_pred_log)
y_test_orig = np.expm1(y_test)

rmse = np.sqrt(mean_squared_error(y_test_orig, y_pred))
mae = mean_absolute_error(y_test_orig, y_pred)
r2 = r2_score(y_test_orig, y_pred)

print('LinearRegression on features:', features)
print(f'RMSE: {rmse:,.2f}   MAE: {mae:,.2f}   R2: {r2:.4f}')
```

```python
from sklearn.model_selection import GridSearchCV, KFold

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', Ridge())
])

param_grid = {
    'model__alpha': [0.1, 1, 10, 50, 100]
}

cv = KFold(n_splits=5, shuffle=True, random_state=42)
grid = GridSearchCV(pipe, param_grid, cv=cv, scoring='neg_root_mean_squared_error')
grid.fit(X, y_log)

print('Best Ridge params:', grid.best_params_)
print('Best CV RMSE (log-space):', -grid.best_score_)
```

```python
y_pred_log = lr.predict(X_test_scaled)
residuals = y_test - y_pred_log

plt.figure(figsize=(6,4))
plt.scatter(y_pred_log, residuals, alpha=0.5)
plt.axhline(0, color='k', linewidth=1)
plt.xlabel('Predicted log(SalePrice)')
plt.ylabel('Residuals (actual - pred)')
plt.title('Residual plot')
plt.show()
```
